```python
# Ilham Hadi Bin Shamsul IS01081469
# Azfar Nazrin Bin Affandi IS01081493

import re
import json
import requests
from datetime import datetime
import csv

url = 'https://shopee.com.my/%E3%80%90South-Korea%E3%80%91korea-sim-card-Korea-Travel-Sim-Card-Unlimited-Data%E3%80%903-30-days%E3%80%91Korea-Travel-Prepaid-Sim-Card-%E9%9F%A9%E5%9B%BD%E4%B8%8A%E7%BD%91%E5%8D%A1-i.976514375.15897114048?sp_atk=bfd9226c-1235-45bf-834f-5a2a6fa0b60a&xptdk=bfd9226c-1235-45bf-834f-5a2a6fa0b60a'

# Extracting shop ID,item ID from the URL using regular expressions
r = re.search(r'i\.(\d+)\.(\d+)', url)
if r:
    shop_id, item_id = r.group(1), r.group(2)
else:
    print("Invalid URL")
    exit()

offset = 0

# File name for the CSV file to store the extracted data
csv_filename = "shopee_reviews.csv"

# Open the CSV file in write mode and specify the header row
with open(csv_filename, mode='w', newline='', encoding='utf-8') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Username', 'Date', 'Content'])

    # Loop to fetch reviews in batches of 20 until all reviews are fetched
    while True:
        # Construct the URL to fetch ratings data
        ratings_url = f'https://shopee.com.my/api/v2/item/get_ratings?filter=0&flag=1&itemid={item_id}&limit=20&offset={offset}&shopid={shop_id}&type=0'
        # Fetch ratings data from the Shopee API
        data = requests.get(ratings_url).json()

        # Iterate through each rating and write it to the CSV file
        for rating in data['data']['ratings']:
            # Extract username, date, and content of the review
            username = rating['author_username']
            date = datetime.fromtimestamp(rating['ctime']).strftime('%Y-%m-%d %H:%M:%S')  # Convert timestamp to human-readable date format
            content = rating['comment']

            # Write the review data to the CSV file
            writer.writerow([username, date, content])

        # Check if there are fewer than 20 ratings fetched, indicating the end of reviews
        if len(data['data']['ratings']) < 20:
            break

        # Increment the offset for the next batch of ratings
        offset += 20

print("Data saved to", csv_filename)
```
